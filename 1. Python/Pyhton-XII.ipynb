```python
def sq(n):
  for i in range(n):
    return i**2
```

```python
sq(3)
```

```python
def sq(n):
  for i in range(n):
    yield i**2
```

```python
sq(2)
```

```python
k = sq(6)
```

```python
for i in k:
  print(i)
```

```python
next(k)
```

```python
#new function

def new():
  n = 1
  print("line one")
  yield n
  
  n = n+1
  print("2nd line")
  yield n

  n = n-1
  print("thrid")
  yield n
```

```python
for i in new():
  print(i)
```
