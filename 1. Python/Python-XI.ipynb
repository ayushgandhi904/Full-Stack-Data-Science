```python
def div(x, y):
  try:
    print(x/y)
  except ZeroDivisionError as e:
    print(e)
```

```python
div(5, 2)
div(2, 0)
div(10, 4)
```

```python
def div(x, y):
  try:
    print(x/y)
  except ZeroDivisionError as e:
    print(e)
  else:
    print("you were correct")
  finally:
    print("you reached")
```

```python
div(5, 2)
div(2, 0)
div(10, 4)
```

```python
x = 10
y = 0

try:
  print("Out")
  try:
    print("nested")
    print(x/y)
  except TypeError as t:
    print(t)
except ZeroDivisionError as z:
  print("outer except")
  print(z)
```
