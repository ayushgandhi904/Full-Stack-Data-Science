```python
def time(func):
  def test():
    import time
    print("exceution")
    start = time.time()
    func()
    end = time.time()
    print(end - start)
  return test

  
```

```python
@time
def ayush():
  """this is function which will take a decorator for calculating a exceution timeing of a function """
  l  = [1, 2, 3, 4, 5]
  l1 = []
  for i in l:
    if i%2 == 0:
      l1.append(i)
  return l1
```
