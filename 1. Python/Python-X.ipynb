```python
def dec(func):
  def test(*args, **kwargs):
    func(*args, **kwargs)
    print(func(*args, **kwargs))
    print("this is my dec func")
    return func(*args, **kwargs)
  return test
```

```python
@dec
def test1(a, b, c, d):
  return a+b-c+d
```

```python
test1(19, 19.2, 19, 3)
```

```python
@dec
def testk(**kwargs):
  return kwargs
```

```python
testk(a = 3, b = -2, c = "hello")
```
