### AND & OR statement

```python
#AND - if both true then true else false
```

```python
vegetable = True
salt = True

if vegetable and salt:
  print("Good dish")
else:
  print("Bad dish")
```

```python
## OR - if both false then false else true
```

```python
car = True
bike = False

if car or bike:
  print("I can travel")
else:
  print("I can't travel")
```

```python
car = False
bike = False

if car or bike:
  print("I can travel")
else:
  print("I can't travel")
```

```python
#to practiacl use

attendance = 70
assignment = 40

if attendance >= 30 and assignment > 50:
  print("You are eligiable ")
else:
  print("Not eligible")
```
