## Type Convertor
- We can convert one data type to another if it is allowed

```python
#string to integer
A = "50.22"
print(f"Before type {type(A)}")
A = float(A)
print(f"After type {type(A)}")
```

```python
#float to integer
B = 92.3
print(f"Before: {type(B)}")
B = int(B)
print(f"After: {type(B)}")
```

```python
B
```

```python
#inter into booleand
A = 0
print(f"Data type:{type(A)}, value: {A}")
A = bool(A)
print(f"Data type: {type(A)}, value: {A}")
```

```python
#inter into booleand
A = -5
print(f"Data type:{type(A)}, value: {A}")
A = bool(A)
print(f"Data type: {type(A)}, value: {A}")
```

```python
#for 0 it will be False, rest all boolean it wil True
#for empty string it will be False, rest value it will be True
A = ""
print(f"Data type:{type(A)}, value: {A}")
A = bool(A)
print(f"Data type: {type(A)}, value: {A}")
```
