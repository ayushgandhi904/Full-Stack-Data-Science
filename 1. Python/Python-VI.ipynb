```python
#user input 

my_list = list()
n = int(input("Enter the number of items you have to purchased: "))
for i in range(n):
  user = input("Enter the item you want to enter:")
  my_list = my_list + [user]
print(my_list)
```

```python
#using while loop 

my_list = list()
n = int(input("Enter the number of items you have to purchased: "))
i = 0
while i<n:
  user = input("Enter the item you want to enter:")
  my_list = my_list + [user]
  i += 1
print(my_list)
```

```python
"en" in my_list
```

```python
if 24 in my_list:
  print("Already in list")
else:
  print("NOt in list, it will need to be added")
  my_list = [24] + my_list
print(my_list)
```
