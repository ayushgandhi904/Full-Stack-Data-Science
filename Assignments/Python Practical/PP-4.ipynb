### Q-1 Write a Python Program to find the factorial of a number.

```python
def fac(a):
    i = a
    a = 1
    while i > 0:
        a = a*i
        i = i-1
    print(a)
```

```python
fac(5)
```

### Q-6 Write a Python Program to find the sum of natural numbers.

```python
def sum(a, b):
    if a, b < 0:
        return "Not the natural numbers"
    else:
        return a+b
```
